```python
import pandas as pd
import json

with open('/Users/heng-tsertsai/.secret/yelp_api.json') as f:
    login = json.load(f)
```

```python
# import the YelpAPI Class
from yelpapi import YelpAPI
# Create an instance with your key
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Davenport, IA',
                                       term='Chinese')
print(type(search_results))
search_results.keys()
```

```python
search_results['total']
```

```python
business = pd.DataFrame(search_results['businesses'])
business.head() 
```

```python
# add offset to our original api call
search_results = yelp_api.search_query(location='Davenport, IA',
                                       term='Chinese',
                                       offset = 20)
```

```python
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head()
```

```python
## concatenate the previous results and new results. 
businesses = pd.concat([business, biz20],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```
